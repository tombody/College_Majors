# The employment outcomes of American College students between 2010 and 2012

The original data was collected by the American Community Survey and was cleaned and realeased by FiveThirtyEight.

Each row in the dataset represents a different major. Some of the columns of the data are described below.

- `Rank` - Rank by median earnings.
- `Major` - Major description.
- `Total` - Total number of people with that major.
- `Sample_size` - The sample size of full-time workers.
- `ShareWomen` - Women as a share of the total.
- `Employed`- The number of employed students
- `Median` - Median salary of full-time, year-round workers.
- `Low_wage_jobs` - The number of students in low wage service jobs.
- `Full_time` - The number of students employed 35 hours or more.
- `Part_time` - The number of students employed less than 35 hours.

```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
recent_grads = pd.read_csv("recent-grads.csv")
print(recent_grads.describe)
#print(recent_grads.iloc[0])
#print(recent_grads.head())
#print(recent_grads.tail())

```

```python
# A quick visual analysis of the data 

recent_grads.plot(x='Sample_size', y='Median', kind='scatter', title='Sample_size vs Median')
recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter', title='Sample_size vs Unemployment_rate')
recent_grads.plot(x='Full_time', y='Median', kind='scatter', title='Full_time vs Median')
recent_grads.plot(x='ShareWomen', y='Median', kind='scatter', title='ShareWomen vs Median')
recent_grads.plot(x='Men', y='Median', kind='scatter', title='Men vs Median')
recent_grads.plot(x='Women', y='Median', kind='scatter', title='Women vs Median')
```

```python
cols = ['Sample_size','Median', 'Employed', 
'Full_time', 'ShareWomen', 'Unemployment_rate', 
'Men', 'Women']

fig = plt.figure(figsize=(5,12))

for r in range(1,8):
    ax = fig.add_subplot(8,1,r)
    ax = recent_grads[cols[r]].plot(kind='hist')
```
